```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('suicide-rates-overview-1985-to-2016.zip')
df
```

```python
# get countries that start with U
u_countries = df[df['country'].str.startswith('U')]
u_countries
```

```python
# get all the unique countries

u_countries.country.unique()
```

```python
# get USA rows

usa_df = u_countries[u_countries['country'] == 'United States']
usa_df
```

```python
# Group by age column and sum # suicides

usa_df.groupby('age').sum()[['suicides_no']]
```

```python
# Change the age column to a categorical one

usa_df['age'] = usa_df['age'].astype('category')
usa_df.info()

cats = usa_df['age'].cat.categories
cats

```

```python
# Re-order the categories so that 5-14 years appears first 

usa_df['age'].cat.reorder_categories([cats[3], cats[0], cats[1], cats[2], cats[4], cats[5]], inplace=True)
usa_df['age'].cat.categories
```

```python
# Draw a barplot of # suicides(y) per age-group (x) for 2015

x = usa_df['age'].cat.categories
y = usa_df.groupby('age').sum()['suicides_no']

plt.figure(figsize=(20,12));
plt.xlabel('Age')
plt.ylabel('Suicides')
plt.title('Age vs. Suicide Rate')
plt.bar(x, y)
plt.savefig("age_suicide.pdf")
```
