```python
import pandas as pd
import numpy as np
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('regrex1.csv')
data.head()
```

```python
data.info()
```

```python
data.shape
```

```python
# Just looking at the data
x = data['x']
y = data['y']
```

```python
plt.figure(figsize=(10,10))
sns.distplot(x, color='orange')
sns.distplot(y);
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(x, y, hue=y);
```

```python
# Formula
# First variable we put in the (f) formula is the dependent variable...in this case it is 'y'
# .fit() gives us the fitted line
f = 'y ~ x'
reg = ols(f, data=data).fit()
```

```python
reg.summary()
```

```python
plt.figure(figsize=(10,10))
sns.lmplot('x', 'y', data=data);
```

```python
plt.figure(figsize=(10,10))
sns.regplot('x', 'y', data=data)
```

```python
preds = reg.predict()
preds
```
